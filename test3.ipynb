```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("HR.csv")
```

```python
f = plt.figure()
f.add_subplot(1,3,1)
sns.distplot(df["satisfaction_level"],bins=10,kde=False)#hist=True
plt.show()
```

```python
f = plt.figure()
f.add_subplot(1,3,1)
sns.histplot(df["satisfaction_level"],bins=10)#hist=True
```

```python
# add another subplot
f.add_subplot(1,3,2)
sns.histplot(df["last_evaluation"],bins=10)
```
