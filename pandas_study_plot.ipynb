```python
## import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 绘制线图

```python
## draw
# with Series
s= pd.Series(np.random.randn(10).cumsum(0), 
             index =np.arange(0,100,10))  #创建Series随机数
s
```

```python
s.plot()
```

```python
# with DataFrame
df = pd.DataFrame(np.random.randn(10,4).cumsum(0), columns = ['A','B','C','D'], index = np.arange(0,100,10)) #创建dataFrame对象
df
```

```python
df.plot()
```

### 绘制柱形图

```python
#柱形图
#处理Series
fig, axes = plt.subplots(2,1)	#获得figure和Axes对象,2,1 means numrows,numcols
```

```python
axes
```

```python
fig
```

```python
data = pd.Series(np.random.rand(16),index = list('abcdefghijklmnop'))  
#创建Series
data
```

```python
data.plot(kind='bar', ax = axes[0],color='k',alpha=0.7)    
#垂直的柱形图,利用kind属性='bar'
```

```python
data.plot(kind='barh', ax = axes[1],color='k',alpha=0.7)   #水平的柱形图,利用kind属性='barh'
```

```python
fig
```

```python
#处理DataFrame
df = pd.DataFrame(np.random.rand(6,4),index = ['one','two','three','four','five','six'],
               columns = pd.Index(['A','B','C','D'],name = 'Genus'))    
#创建DataFrame
df.plot(kind='bar') #垂直的柱形图,利用kind属性='bar'
```

### 直方图与密度图

```python
#直方图和密度图
plt.figure()
data = pd.Series(np.random.rand(16),index = list('abcdefghijklmnop'))  #创建Series
data.hist(bins = 50)    #直方图，bins参数代表有多少个区间
data.plot(kind='kde')  #密度图，将kind属性设置为kde
##使用同一张画布，两个图会在一起
plt.show()
```

### 子图

```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline

from datetime import datetime

tech_stocks_data = pd.read_csv('https://raw.githubusercontent.com/nicholasmccullum/python-visualization/master/tech_stocks/GOOG_MSFT_FB_AMZN_data.csv')

tech_stocks_data.sort_values('Period', ascending = True, inplace = True)

google = tech_stocks_data['Alphabet Inc Price']

amazon = tech_stocks_data['Amazon.com Inc Price']

facebook = tech_stocks_data['Facebook Inc Price']

microsoft = tech_stocks_data['Microsoft Corp Price']

dates = tech_stocks_data['Period']

x = []

for date in tech_stocks_data['Period']:

    x.append(datetime.strptime(date, '%Y-%m-%d %H:%M:%S').year)

    

plt.figure(figsize=(16,12))

#Plot 1

plt.subplot(2,2,1)

plt.xticks(np.arange(0, len(x) + 1)[::365], x[::365])

plt.plot(dates, google)

plt.title('Alphabet (GOOG) (GOOGL) Stock Price')

#Plot 2

plt.subplot(2,2,2)

plt.xticks(np.arange(0, len(x) + 1)[::365], x[::365])

plt.plot(dates, amazon)

plt.title('Amazon (AMZN)) Stock Price')

#Plot 3

plt.subplot(2,2,3)

plt.xticks(np.arange(0, len(x) + 1)[::365], x[::365])

plt.plot(dates, facebook)

plt.title('Facebook (FB) Stock Price')

#Plot 4

plt.subplot(2,2,4)

plt.xticks(np.arange(0, len(x) + 1)[::365], x[::365])

plt.plot(dates, microsoft)

plt.title('Microsoft (MSFT) Stock Price')
```
